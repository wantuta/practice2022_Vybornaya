```python
import numpy as np

# Пункт 1: Чтение данных
df_sample = np.genfromtxt("sample.csv", delimiter=",")

# Пункт 2: Вычисление среднего значения и стандартного отклонения для каждой строки
data_mean = np.mean(df_sample, axis=1)
data_std = np.std(df_sample, axis=1)

# Вывод среднего значения и стандартного отклонения
for index, row in enumerate(df_sample):
    print(f'{index}\t {row.mean()}\t {row.std()}')

# Пункт 3: Переименование строк
renamed_rows_df = np.column_stack((np.arange(2, len(df_sample) + 2), data_mean, data_std))

# Пункт 4: Фильтрация строк со стандартным отклонением меньше 250
std_less_250_df = renamed_rows_df[renamed_rows_df[:, 2] < 250]

# Запись в файл
np.savetxt('output.txt', std_less_250_df, delimiter='\t', fmt='%.2f')
```
